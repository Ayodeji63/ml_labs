```python
import numpy as np

X = np.arange(-2, 3) # 
Y = np.arange(-2, 3) #
X, Y = np.meshgrid(X, Y)
```

```python

Z = X**2 + Y**2
print(Z)
```

```python
import matplotlib.pyplot as plt

x = np.arange(-3, 3, 0.5)
y = np.arange(-3, 3, 0.5)

X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2)) # ripples/wave function

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.show()
```
